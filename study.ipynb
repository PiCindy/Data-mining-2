```python
ind = context_df['Unnamed: 0'].values
```

```python
transactions = []

for i in ind:
    cols = []
    if i[-3:] == 'ear':
        continue
    if i[-3:] == 'pro':
        cols.append('pro')
    else:
        cols.append('con')
    for c in context_df.columns[1:]:
        if context_df[context_df['Unnamed: 0'] == i][c].values[0] == 1:
            cols.append(c)
    transactions.append(cols)
```

```python
from apyori import apriori
```

```python
results = list(apriori(transactions))
```

```python
results
```

```python
for r in results:
    if 'pro' in r.items or 'con' in r.items:
        for stat in r.ordered_statistics:
            if stat.confidence > 0.7 and 'con' in stat.items_add:
                print(stat.items_base, stat.items_add, stat)
```

```python
for r in results:
    #if len(r.items) < 4:
    #    continue
    if 'pro' in r.items or 'con' in r.items:
        for stat in r.ordered_statistics:
            if len(stat.items_base) < 3:
                continue
            if stat.confidence > 0.625 and 'pro' in stat.items_add:
                print(stat.items_base, stat.items_add, stat)
```

```python
all_rules = []
for r in results:
    if 'pro' in r.items or 'con' in r.items:
        for stat in r.ordered_statistics:
            if len(stat.items_add) == 1 and ('pro' in stat.items_add or 'con' in stat.items_add):
                all_rules.append((r.support, stat))
```

```python
all_rules.sort(key=lambda x: -x[1].confidence)
```

```python
for rule in all_rules:
    print(rule)
```
